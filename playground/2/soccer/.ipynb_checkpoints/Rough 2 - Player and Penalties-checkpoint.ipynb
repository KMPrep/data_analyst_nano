```python
import pandas as pd

query = 'select \
P.Player_api_id as player_api_id \
,P.player_name as player_name \
,PA.date as rating_date \
,PA.penalties as penalties \
from Player as P \
inner join Player_Attributes as PA \
on P.player_api_id = PA.player_api_id'

print(query)
```

```python
import sqlite3 as db

conn = db.connect("database.sqlite")
df = pd.read_sql_query(query, conn)
df.head(5)


```

```python
df.columns.values
```

```python
df.info()
```

```python
df['rating_date'] = pd.to_datetime( df['rating_date'])
df.dropna(inplace=True)
```

```python
df.info()
```

```python
sum(df.duplicated())
```

```python
df['rating_year'] = pd.DatetimeIndex(df['rating_date']).year
df
```

```python
df['Ranking'] = df.sort_values(['player_api_id','rating_date'], ascending=[True,False]) \
                  .groupby(['player_api_id','rating_year']) \
                  .cumcount() + 1
df
```

```python
#calculate average penalties for each player for every year
avg_penalties_per_player_per_year = df.groupby(['player_api_id','rating_year']).mean().reset_index()
avg_penalties_per_player_per_year
```

```python
#identify players who been rated atleast for last 5 years (2011 - 2016)

max_year =  avg_penalties_per_player_per_year['rating_year'].max() 
print(max_year.dtype)
start_year = max_year - 5

avg_penalties_per_player_per_year = avg_penalties_per_player_per_year[ ((avg_penalties_per_player_per_year['rating_year'] >= start_year) \
                & (avg_penalties_per_player_per_year['rating_year'] <= max_year)) ]

avg_penalties_per_player_per_year

#calculate penalties average for all players for all years

avg_penalties_per_player_for_all_years = avg_penalties_per_player_per_year.groupby(['player_api_id']).mean().reset_index()
interested_cols =  avg_penalties_per_player_for_all_years.iloc[:, [0,2]] 
player_with_high_penalties_rating = interested_cols.sort_values(by='penalties', ascending=False).head(1).iloc[0,[0]]
player_id =  player_with_high_penalties_rating.mean()
```
