```python
#which team attributes lead to most victories

# find the team with most victories in all the season and compare that data with team attributes


import pandas as pd

query = 'select \
M.id as match_id \
,M.date as match_date \
,L.name as league_name \
,M.stage as match_stage \
,C.name as country_name \
,M.season as season \
,HT.team_long_name as home_team_long_name \
,HT.team_short_name as home_team_short_name \
,HT.team_api_id as home_team_api_id \
,AT.team_api_id as away_team_api_id \
,AT.team_long_name as away_team_long_name \
,AT.team_short_name as away_team_short_name \
,M.home_team_goal as home_team_goals \
,M.away_team_goal as away_team_goals \
from Match as M \
join League as L on M.league_id = L.id \
join Country as C on C.id = L.country_id \
join Team as HT on HT.team_api_id = M.home_team_api_id \
join Team as AT on AT.team_api_id = M.away_team_api_id'

print(query)

team_attr_query = 'select \
T.team_api_id as team_api_id \
, T.team_long_name as team_long_name \
, T.team_short_name as team_short_name \
, TA.date as rating_date \
, TA.buildUpPlaySpeed as buildup_speed \
, TA.buildUpPlaySpeedClass as buildup_speed_class \
, TA.buildUpPlayDribbling as buildup_play_dribbling \
, TA.buildUpPlayDribblingClass as buildup_play_dribbling_class \
, TA.buildUpPlayPassing as buildup_play_passing \
, TA.buildUpPlayPassingClass as buildup_play_passing_class \
, TA.buildUpPlayPositioningClass as buildup_play_positioning_class \
, TA.chanceCreationPassing as chance_creation_passing \
, TA.chanceCreationPassingClass as chance_creation_passing_class \
, TA.chanceCreationCrossing as chance_creation_crossing \
, TA.chanceCreationCrossingClass as chance_creation_crossing_class \
, TA.chanceCreationShooting as chance_creation_shooting \
, TA.chanceCreationShootingClass as chance_creation_shooting_class \
, TA.chanceCreationPositioningClass as chance_creation_positioning_class \
, TA.defencePressure as defence_pressure \
, TA.defencePressureClass as defence_pressure_class \
, TA.defenceAggression  as defence_aggression \
, TA.defenceAggressionClass as defence_aggresion_class \
, TA.defenceTeamWidth as defence_team_width \
, TA.defenceTeamWidthClass as defence_width_class \
, TA.defenceDefenderLineClass as defence_defender_line_class \
from Team as T \
join Team_Attributes as TA \
on T.team_api_id = TA.team_api_id'
```

```python
def get_winning_team(df):
    if df['home_team_goals'] > df['away_team_goals'] :
        return df['home_team_api_id']
    elif df['home_team_goals'] < df['away_team_goals']:
        return df['away_team_api_id']
    else:
        return 'No result'

df['winning_team'] = df.apply(get_winning_team, axis=1)
```

```python
df.head(10)
```

```python
#In every league/season, find the team who wins most matches

df_winning_team = df.query('winning_team != "No result"')
df_winning_team.head(5)

df_winning_team_grouped_data = df_winning_team.groupby(['league_name', 'season', 'winning_team']).count()['match_id'].reset_index()
df_winning_team_grouped_data.rename(index=str, columns={"match_id": "win_count"}, inplace=True)
df_winning_team_grouped_data

df_winning_team_grouped_data['ranking'] = df_winning_team_grouped_data.sort_values(['league_name','season', 'win_count', 'winning_team'], ascending=[True,True,False,False]) \
                                              .groupby(['league_name','season']) \
                                              .cumcount() + 1

df_winning_team_grouped_data = df_winning_team_grouped_data.query('ranking == 1')
```

```python
#Find out which 2 opponents always ends in draw.

def hash_data(ra):
    p = ','.join(map(str, ra))
    return hashlib.md5(p).hexdigest()

def hash_teams(df):
    teams = []
    teams.append(df['home_team_api_id'])
    teams.append(df['away_team_api_id'])
    teams.sort()
    p = ','.join(map(str, teams))
    return p

df['team_hash'] = df.apply(hash_teams, axis=1)

```

```python
df.head(10)
```

```python
#matches between two teams

total_matches_between_teams = df.groupby('team_hash').count()['match_id'].reset_index()
total_matches_between_teams.rename(index=str, columns={'match_id': 'total_match_count'}, inplace=True)
total_matches_between_teams
```

```python
df_draw_matches =  df.query('winning_team == "No result"')
df_draw_matches_by_teams = df_draw_matches.groupby('team_hash').count()['match_id'].reset_index()
df_draw_matches_by_teams.rename(index=str, columns={'match_id': 'draw_match_count'}, inplace=True)
df_draw_matches_by_teams.sort_values(by='draw_match_count', ascending=False)
```

```python
total_and_draw_queries = pd.merge(total_matches_between_teams,df_draw_matches_by_teams, how='inner', left_on='team_hash', right_on='team_hash')
total_and_draw_queries['draw_match_proportions'] = total_and_draw_queries['draw_match_count'] / total_and_draw_queries['total_match_count']
total_and_draw_queries.head(10)

total_and_draw_queries.sort_values(by='draw_match_proportions', ascending=False)
total_and_draw_queries.describe()
```

```python
#Find the mean of 'total_match_count', so that, we dont include matches between teams who have played very less

total_and_draw_queries = total_and_draw_queries.query('total_match_count > 8')
total_and_draw_queries.sort_values(by='draw_match_proportions', ascending=False)
```
