**XML Data Convertion**

Generic XML Parser Class is created to convert XML to Dataframe to get meaningful insights.

```python
"""
The below code is copied from here: 
http://www.austintaylor.io/lxml/python/pandas/xml/dataframe/2016/07/08/convert-xml-to-pandas-dataframe/
"""

class XML2DataFrame:

    def __init__(self, xml_data):
        self.root = ET.XML(xml_data)

    def parse_root(self, root):
        """Return a list of dictionaries from the text
         and attributes of the children under this XML root."""
        return [self.parse_element(child) for child in iter(root)]

    def parse_element(self, element, parsed=None):
        """ Collect {key:attribute} and {tag:text} from thie XML
        element and all its children into a single dictionary of strings."""
        
        if parsed is None:
            parsed = dict()
        for key in element.keys():
            parsed[key] = element.attrib.get(key)
        if element.text:
            parsed[element.tag] = element.text
            
        """ Apply recursion"""
        for child in list(element):
            self.parse_element(child, parsed)
        return parsed

    def process_data(self):
        """ Initiate the root XML, parse it, and return a dataframe"""
        structure_data = self.parse_root(self.root)
        return pd.DataFrame(structure_data)
```

Since each attribute has XML formats in a unique way, a generic function is needed to handle each attribute.

```python
""" This function takes the dataframe and a name of the column to be parsed"""

def parse_xml_data(dataFrame, colName=None):    
    
    """ Handle Empty Cells"""
    if dataFrame[colName] is None:
        if colName in ['corner', 'shoton', 'shotoff', 'foulcommit', 'possession']:
            return pd.Series([None, None])
        elif colName in ['card']:
            return pd.Series([None, None, None, None])
        else:
            raise ValueError('Unknown column to parse') 
    
    """ Get the data from the column and convert it to Dataframe """    
    xml2df = XML2DataFrame(dataFrame[colName])
    xml_dataframe = xml2df.process_data()

    """ Gather the Home and Away ID """
    home_team_api_id = dataFrame['home_team_api_id']
    away_team_api_id = dataFrame['away_team_api_id']                    

    """ Process 'corner', 'shoton', 'shotoff', 'foulcommit' columns"""
    if colName in ['corner', 'shoton', 'shotoff', 'foulcommit']:
        """ Initialize the return data """
        series = pd.Series([None, None])
        try:  
            """ Check if team column is in parsed dataframe """
            if 'team' in xml_dataframe.columns:
                data_values = xml_dataframe.groupby('team').count()['id'].reset_index()
                data_values['team'] = data_values['team'].astype(int) 

                home_team_data_value = data_values[data_values['team'] == home_team_api_id]['id']
                away_team_data_value = data_values[data_values['team'] == away_team_api_id]['id']
                series =  pd.Series([  0 if home_team_data_value.count() == 0 else home_team_data_value.iloc[0],      
                                     0 if away_team_data_value.count() == 0 else away_team_data_value.iloc[0]])                
        except:
            print('{0} : {1} Problematic Index'.format( colName, dataFrame['match_id']))
        return series
    elif colName == 'possession':
        series = pd.Series([None, None])
        try:
            if set(xml_dataframe.columns).issuperset(set(['awaypos','homepos', 'elapsed'])) :  
                ### There could be data where elapsed time is not 90.. 
                ### We need to highest elapsed and parse the data.
                """ 
                During some data exploration phase, we have identified, that some matches possession data is not available for the entire.
                In the below logic, we try to sort the possession data available for the highest elapsed time and get their data.    
                """
                xml_dataframe['elapsed']  = xml_dataframe['elapsed'].astype(int)
                row_with_max_elapsed = xml_dataframe['elapsed'].idxmax()                    
                homepossesion_time = xml_dataframe.iloc[row_with_max_elapsed]['homepos'] 
                awaypossesion_time = xml_dataframe.iloc[row_with_max_elapsed]['awaypos']

                series = pd.Series([int(homepossesion_time), int(awaypossesion_time)])   

        except Exception as e:                
                return series
        return series
    elif colName == 'card':            
        """ For card column, we would like to get the data in this format 
        ('home team yellow card', 'away team yellow card', 'home team red card', 'away team red card' )"""
        series = pd.Series([None, None, None, None])

        try:                
            if set(xml_dataframe.columns).issuperset(set(['team', 'ycards'])):                    
                yCard_frame = xml_dataframe.groupby(['team','ycards']).count()['id'].reset_index() 
                yCard_frame['team'] = yCard_frame['team'].astype(int) 

                home_team_data_value = yCard_frame[(yCard_frame['team'] == home_team_api_id) & (yCard_frame['ycards'] == '1')]['id']
                away_team_data_value = yCard_frame[(yCard_frame['team'] == away_team_api_id) & (yCard_frame['ycards'] == '1')]['id']                   

                series[0] = 0 if home_team_data_value.count() == 0 else home_team_data_value.iloc[0]
                series[1] = 0 if away_team_data_value.count() == 0 else away_team_data_value.iloc[0]

                #update red card 0 for now, rather than None, as red card is not issued in all the matches.
                series[2] = 0
                series[3] = 0

            if set(xml_dataframe.columns).issuperset(set(['team', 'rcards'])):

                rCard_frame = xml_dataframe.groupby(['team','rcards']).count()['id'].reset_index()
                rCard_frame['team'] = rCard_frame['team'].astype(int) 
                home_team_data_value = rCard_frame[(rCard_frame['team'] == home_team_api_id) & (rCard_frame['rcards'] == '1')]['id']
                away_team_data_value = rCard_frame[(rCard_frame['team'] == away_team_api_id) & (rCard_frame['rcards'] == '1')]['id']                   
                series[2] = 0 if home_team_data_value.count() == 0 else home_team_data_value.iloc[0]
                series[3] = 0 if away_team_data_value.count() == 0 else away_team_data_value.iloc[0]                 

            return series

        except:
            print('{0} : {1} Problematic Index'.format( colName, dataFrame['match_id']))
            return series
    else:
        raise ValueError('Unknown column to parse') 
```

```python
""" Now parse all the XML data columns and assign it to new column"""
#shoton
df_match[['home_team_shoton','away_team_shoton']] = df_match.apply(parse_xml_data, colName='shoton', axis = 1)
#shotoff
df_match[['home_team_shotoff','away_team_shotoff']] = df_match.apply(parse_xml_data, colName='shotoff', axis = 1)
#corners
df_match[['home_team_corners','away_team_corners']] = df_match.apply(parse_xml_data, colName='corner', axis = 1)
#foulcommits
df_match[['home_team_foulcommit','away_team_foulcommit']] = df_match.apply(parse_xml_data, colName='foulcommit', axis = 1)
#possession
df_match[['home_team_possesion','away_team_possession']] = df_match.apply(parse_xml_data, colName='possession', axis = 1)
#card - yellow and red cards
df_match[['home_team_ycard','away_team_ycard','home_team_rcard','away_team_rcard']] = df_match.apply(parse_xml_data, colName='card', axis = 1)
```

```python
""" This function takes Numerical Column and Categorized Column to find Min and Max Range"""

def find_min_max(data_frame = None, numerical_col_name =None, categorized_col_name = None):
    df_filtered = data_frame.loc[:,[numerical_col_name,categorized_col_name]]   
    df_group_min = df_filtered.groupby(categorized_col_name).min().reset_index()    
    df_group_max = df_filtered.groupby(categorized_col_name).max().reset_index()

    return df_group_min.merge(df_group_max, how='inner', left_on=categorized_col_name, right_on=categorized_col_name)

find_min_max(data_frame = df_team, numerical_col_name = 'buildup_speed', categorized_col_name = 'buildup_speed_class' )
```

### Question 1 - Which team have improved over a period of time?

This is a broad question, as team's improvement could be measured by matches won, goals scored, foul commits and cards issued.

For now, Let's look at from matches won. For that, the dataset we have is: 'goals scored by each team (i.e. home_team_goals, away_team_goals)'. So, we need to identify the winning team.

```python
""" Let's filter the data frame to include only necessary columns """
win_match_cols = ['match_id', 'league_id', 'league_name', 'season', 'season_start','season_end', 'home_team_api_id', 'home_team_short_name', 'home_team_long_name'\
                 ,'away_team_api_id', 'away_team_short_name', 'away_team_long_name','home_team_goals','away_team_goals']

df_match_win_matches_filtered = df_match.loc[:,win_match_cols]
df_match_win_matches_filtered.head(5)
```

```python
""" determine the winning team by comparing the goals """
def get_winning_team(df):
    if df['home_team_goals'] > df['away_team_goals'] :
        return str(df['home_team_api_id'])
    elif df['home_team_goals'] < df['away_team_goals']:
        return str( df['away_team_api_id'])
    else:
        return 'No result'

df_match_win_matches_filtered['winning_team'] = df_match_win_matches_filtered.apply(get_winning_team, axis=1)
df_match_win_matches_filtered.head(5)
```

Now we have determined the winning team, lets find total matches played the team every season, and total matches won the team every season

```python
""" determine the matches played all the teams """

def get_matches_played_by_each_team(data_frame = None, league_id = None, season_start=None):
    
    data_frame_filtered = data_frame
    """ Filter the Data Frame if league id is present"""
    if league_id is not None:
        data_frame_filtered = data_frame_filtered.query("league_id == @league_id")
    """Filter the data frame if season is provided """   
    if season_start is not None:
        data_frame_filtered = data_frame_filtered.query("season_start == @season_start")
    
    """home matches played by the team"""
    home_games_played_team = data_frame_filtered.groupby(['season_start', 'home_team_api_id']).count()['match_id'].reset_index()
    home_games_played_team.rename(index=str, columns={"match_id": "home_match_count"}, inplace=True)
    
    """away matches played by the team"""
    away_games_played_team = data_frame_filtered.groupby(['season_start', 'away_team_api_id']).count()['match_id'].reset_index()
    away_games_played_team.rename(index=str, columns={"match_id": "away_match_count"}, inplace=True)
    
    """ Join both away and home games to get the total matches"""
    all_matches_played_team = pd.merge(home_games_played_team, away_games_played_team,  how='inner', \
                                        left_on=['season_start','home_team_api_id'], right_on = ['season_start','away_team_api_id'])
    
    """ Clean Column Names"""
    all_matches_played_team.rename(index=str, columns={"home_team_api_id": "team_api_id"}, inplace=True)
    all_matches_played_team.drop('away_team_api_id',axis=1, inplace=True)
    
    """Compute Total Matches"""
    all_matches_played_team['total_match_count'] = all_matches_played_team['home_match_count'] \
                                                    + all_matches_played_team['away_match_count']
    return all_matches_played_team

""" Testing the function with a sample league id"""
total_matches_played_by_team = get_matches_played_by_each_team(df_match_win_matches_filtered, league_id = 1)
total_matches_played_by_team.head(5)
```

```python
""" testing the function with both league id and season Start """
total_matches_played_by_team = get_matches_played_by_each_team(df_match_win_matches_filtered, league_id = 1,season_start=2015)
total_matches_played_by_team.head(5)
```

```python
""" determine matches won by each team """

def get_matches_won_by_teams(data_frame = None, league_id = None, season_start = None):
    data_frame_filtered = data_frame
    """Filter Data Frame if League Id is provided """    
    if league_id is not None:
        data_frame_filtered = data_frame_filtered.query("league_id == @league_id")
    """Filter Data Frame if Season is provided """    
    if season_start is not None:
        data_frame_filtered = data_frame_filtered.query("season_start == @season_start")
        
    """Remove the drawn matches"""
    data_frame_filtered = data_frame_filtered.query("winning_team != 'No result'")
    
    winning_matches = data_frame_filtered.groupby(['season_start', 'winning_team']).count()['match_id'].reset_index()
    winning_matches['winning_team'] = winning_matches['winning_team'].astype(int)
    
    """Fix column Names"""
    winning_matches.rename(index=str, columns={'match_id': 'win_match_count'}, inplace=True)
    return winning_matches

""" Testing the function with League Id"""
total_won_matches_by_teams =  get_matches_won_by_teams(df_match_win_matches_filtered, league_id = 1)
total_won_matches_by_teams.head(5)
```

```python
""" Testing the function with league id and season """
total_won_matches_by_teams =  get_matches_won_by_teams(df_match_win_matches_filtered, league_id = 1, season_start=2015)
total_won_matches_by_teams.head(5)
```

```python
""" This function computes the team who have improved by winnning most matches """

def get_team_improved_over_the_period(data_frame = None,  league_id = None):
    """ Get all matches played by the team in the league """
    all_matches = get_matches_played_by_each_team(data_frame, league_id= league_id)
    
    """ Get all the won matches by the team in the league """
    won_matches = get_matches_won_by_teams(data_frame=data_frame, league_id=league_id)
    
    """Join the played matches and won matches count """
    all_data = pd.merge(all_matches, won_matches, how='inner', left_on=['season_start','team_api_id'], \
                        right_on=['season_start', 'winning_team'])
    """Fix Column Names """
    all_data.drop('winning_team',axis=1, inplace=True)
    
    """ Compute the winning percentage for all the teams in the league """
    all_data['win_percentage' ] =  all_data['win_match_count']/all_data['total_match_count']
    
    """ Reformat the data, so that we can see all the win percentages and season in every column """
    pivoted_data = all_data.pivot(index='team_api_id', columns='season_start', values='win_percentage').reset_index()    
    pivoted_data.dropna(inplace=True)
    pivoted_data.set_index('team_api_id', inplace=True)
    
    """ Get the win percentage in the last season """
    last_col_name = pivoted_data.columns[len(pivoted_data.columns) -1]
    
    """ Get the win percentage in the first season """
    first_col_name = pivoted_data.columns[0]    
    
    """ Compute the win percentage change between the last season and current season """
    pivoted_data['overall_win_percentage_change'] = pivoted_data[last_col_name] - pivoted_data[first_col_name]
    return pivoted_data.sort_values(by='overall_win_percentage_change', ascending=False).head(1).reset_index()

""" Test the function for a given league """
test_df = get_team_improved_over_the_period(df_match_win_matches_filtered, 17642)
test_df.head(5)
```

```python

""" This function find the win percentages for all the teams in all the leagues """
def get_teams_improved_by_league(data_frame=None):
    leagues = data_frame['league_id'].unique()
    teams_improved_list = []
    
    """Loop through the leagues and get all the teams' win percentage in the league """
    for league in leagues:
        team_improvement = get_team_improved_over_the_period(data_frame, league)        
        teams_improved_list.append(team_improvement)
        
    """Combine all the league data """
    all_data = pd.concat(teams_improved_list, ignore_index=True)
    return all_data

""" Here we compute all the leagues and their teams' win percentage """
team_improved_the_most_from_each_league = get_teams_improved_by_league(df_match_win_matches_filtered)
team_improved_the_most_from_each_league
```

```python
"""This function computes popular team attributes followed most of the team """
def show_popular_team_attr(data_frame=None, rating_year=None):
    
    data_frame_filtered = data_frame
    """ Initial the Series """
    empty_data =  np.empty([len(team_attributes_cols)], dtype=object)
    top_team_attr_in_every_class = pd.Series(empty_data, index=team_attributes_cols)
    
    """Filter the data frame if rating year is provided """
    if rating_year is not None:
        data_frame_filtered = data_frame_filtered.query("rating_year == @rating_year")
    
    nrows = 3
    ncols = 4
    plot_index = 1
    plt.figure(figsize=(12,12))
    """ Loop through all the team attribute column and find most used team attribute for every class of team attribute """
    for c in team_attributes_cols:
        plt.subplot(nrows, ncols, plot_index)        
        data_frame_filtered[team_attributes_cols[plot_index -1]].value_counts().plot(kind='bar')
        top_team_attr_in_every_class[c] = df_team[team_attributes_cols[plot_index -1]].value_counts(ascending=False).idxmax()
        plt.title(team_attributes_cols[plot_index -1])
        plt.xlabel('class')
        plt.ylabel('count')
        plot_index = plot_index + 1

    plt.tight_layout()
    plt.show()
    
    return top_team_attr_in_every_class

""" Get the popular team attributes for the year 2015"""
popular_team_attr_2015 = show_popular_team_attr(df_team, rating_year=2015)
popular_team_attr_2015
```

```python

""" This function gets the teams' match count, won match count and attribute they have followed for a given year"""
def get_team_match_win_count_by_team_by_attr(data_frame=None, rating_year=None, team_attr_name=None):
    
    df_team_filtered = data_frame
    
    """ Filter the data frame, if rating year is provided """
    if rating_year is not None:
         df_team_filtered = data_frame.query("rating_year == @rating_year")
    
    """First We need to figure out the teams following different classes of attributes"""   
    df_team_filtered = df_team_filtered.loc[:,['team_api_id', 'rating_year', team_attr_name]]
    
    """Get all the matches played by the team for that year """
    all_matches_played_by_team_by_year = get_matches_played_by_each_team(data_frame=df_match_win_matches_filtered, \
                                                                         season_start=rating_year)
    
    """Get all the won matches for all the team for the given year """
    won_matches_by_team_year = get_matches_won_by_teams(data_frame=df_match_win_matches_filtered, season_start=rating_year)
    
    """ Merge total matches and won matches for all the team """
    merged_df_matches = all_matches_played_by_team_by_year.merge(won_matches_by_team_year, how='inner' \
                                                                                  , left_on=['season_start', 'team_api_id'] \
                                                                                  , right_on=['season_start', 'winning_team'])
    
    """Now, merge the team attributes along with match count and won match count """
    merged_df = df_team_filtered.merge(merged_df_matches, how='inner', left_on=['team_api_id', 'rating_year'] \
                                                         ,right_on=['team_api_id', 'season_start']    )
    """Fetch only the necessary columns """
    merged_df = merged_df.loc[:,['team_api_id', 'rating_year', team_attr_name, 'total_match_count', 'win_match_count']]
    """Get the team count by team attribute they have followed """
    team_count_by_team_attr = merged_df.groupby(team_attr_name)['team_api_id'].count()
    """ Get the match count by team attribute"""
    match_count_by_team_attr = merged_df.groupby(team_attr_name)['total_match_count'].sum()
    
    """ Get the won match count by team attribute """
    win_count_by_team_attr = merged_df.groupby(team_attr_name)['win_match_count'].sum()
    
    """combine all the results for all the team attributes """
    merge_team_attr_with_match_data_df = pd.concat([team_count_by_team_attr, match_count_by_team_attr, \
                                                        win_count_by_team_attr], axis=1)
    """Compute win percentage """
    merge_team_attr_with_match_data_df['win_percentage'] = merge_team_attr_with_match_data_df['win_match_count'] / merge_team_attr_with_match_data_df['total_match_count']
    return merge_team_attr_with_match_data_df
```
